```
#Bai9
L1 = [1, 2, 3, 4, 5, 6]
L2 = [3, 4, 5, 6, 7, 8, 9, 10]

print('Có trong cả A và B:',end=' ')
for i in range(11):
  if i in L1 and i in L2:
    print(i,end=' ')
print()

print('Chỉ có trong A:',end=' ')
for i in range(11):
  if i in L1 and i not in  L2:
    print(i,end=' ')
print()

print('Chỉ có trong B:',end=' ')
for i in range(11):
  if i not in L1 and i in L2:
    print(i,end=' ')
print()

print('Không có trong cả A và B:',end=' ')
for i in range(11):
  if i not in L1 and i not in L2:
    print(i,end=' ')
print()
```
